# Introducción

Luego de completar la preparación de los datos y el análisis exploratorio, se aplica el Análisis de Componentes Principales (PCA).

Esta técnica permite reducir la dimensionalidad del conjunto de datos conservando la mayor cantidad posible de información, facilitando la visualización y el análisis de patrones presentes en los datos.

# Objetivos

## Objetivo General

Aplicar la técnica de Análisis de Componentes Principales (PCA) para reducir la dimensionalidad del conjunto de datos.

## Objetivos específicos

- Seleccionar las variables numéricas.
- Estandarizar los datos.
- Aplicar PCA.
- Analizar la varianza explicada.
- Interpretar las componentes principales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pd.set_option("display.max_columns", None)
```

# Carga del Dataset Limpio

Se utiliza el conjunto de datos procesado durante el Notebook 2.

```python
df = pd.read_json("../data/processed/streaming_users_clean.json")

df.head()
```

# Selección de Variables Numéricas

PCA únicamente trabaja con variables numéricas. Por ello se seleccionan las variables cuantitativas del conjunto de datos.

```python
variables = [
    "age",
    "monthly_watch_time_mins",
    "customer_support_tickets"
]

X = df[variables]

X.head()
```

# Estandarización de los Datos

Antes de aplicar PCA es necesario estandarizar las variables.

Este procedimiento evita que aquellas variables con valores más grandes tengan mayor influencia en el análisis.

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

## ¿Por qué se utiliza StandardScaler?

El algoritmo PCA es sensible a la escala de las variables.

Por este motivo se aplica una estandarización, logrando que todas las variables tengan media igual a cero y desviación estándar igual a uno.

# Aplicación del PCA

```python
pca = PCA()

componentes = pca.fit_transform(X_scaled)
```

# Varianza Explicada

```python
varianza = pd.DataFrame({
    "Componente": range(
        1,
        len(pca.explained_variance_ratio_)+1
    ),
    "Varianza Explicada":
    pca.explained_variance_ratio_
})

varianza
```

## Interpretación

La varianza explicada indica qué porcentaje de la información original es conservado por cada componente principal.

Las primeras componentes suelen concentrar la mayor parte de la información del conjunto de datos.

# Gráfico de Varianza Explicada

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(
        1,
        len(pca.explained_variance_ratio_)+1
    ),
    pca.explained_variance_ratio_,
    marker="o"
)

plt.xlabel("Componente Principal")

plt.ylabel("Varianza Explicada")

plt.title("Varianza explicada por componente")

plt.grid(True)

plt.show()
```

## Interpretación

El gráfico permite observar la contribución de cada componente principal.

Generalmente las dos primeras componentes concentran la mayor parte de la variabilidad del conjunto de datos.

# Representación de las Componentes Principales

```python
plt.figure(figsize=(8,6))

plt.scatter(
    componentes[:,0],
    componentes[:,1]
)

plt.xlabel("PC1")

plt.ylabel("PC2")

plt.title("Proyección sobre las dos primeras componentes")

plt.grid(True)

plt.show()
```

## Interpretación

La representación gráfica facilita la visualización de la distribución de los registros en un espacio bidimensional.

Esto permite identificar agrupamientos o comportamientos similares entre los usuarios.

```python
# Varianza acumulada
varianza["Varianza Acumulada"] = varianza["Varianza Explicada"].cumsum()

varianza
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    varianza["Componente"],
    varianza["Varianza Acumulada"],
    marker="o"
)

plt.title("Varianza acumulada")
plt.xlabel("Componente")
plt.ylabel("Varianza acumulada")

plt.grid(True)
plt.show()
```

# Interpretacion

La varianza acumulada permite determinar cuántas componentes principales son necesarias para conservar la mayor parte de la información original del conjunto de datos.

# Conclusiones

La aplicación del Análisis de Componentes Principales permitió reducir la dimensionalidad del conjunto de datos conservando la mayor parte de la información disponible.

El escalamiento previo garantizó que todas las variables participaran de forma equilibrada en el análisis.

Los resultados obtenidos constituyen una herramienta útil para comprender la estructura general del conjunto de datos y servirán de apoyo para las conclusiones finales del proyecto.
