```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

plt.style.use("ggplot")

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_json("../data/processed/streaming_users_clean.json")
```

##Pregunta 1
## ¿Cómo se distribuyen las edades de los usuarios?


```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["age"],
    bins=15,
    kde=True
)

plt.title("Distribución de edades")

plt.xlabel("Edad")

plt.ylabel("Frecuencia")

plt.show()
```

**Interpretacion**

La mayor concentración de usuarios se encuentra en edades adultas jóvenes. La distribución permite observar el comportamiento general de la variable y detectar posibles valores extremos.

**Conclusion**

La edad constituye una variable relevante para comprender el perfil general de los usuarios analizados.

## Pregunta 2

## ¿Cuál es el plan de suscripción más utilizado?

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="subscription_plan"
)

plt.title("Usuarios por plan")

plt.show()
```

**Interpretación**

Se observa que uno de los planes concentra la mayor cantidad de usuarios, lo que podría indicar una preferencia por una opción de costo intermedio.

##Pregunta 3
## ¿Los usuarios Premium consumen más contenido?

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="subscription_plan",
    y="monthly_watch_time_mins"
)

plt.show()
```

##Pregunta 4

## ¿Existen diferencias entre países?

```python
plt.figure(figsize=(10,5))

sns.barplot(
    data=df,
    x="country",
    y="monthly_watch_time_mins"
)

plt.xticks(rotation=45)

plt.show()
```

##Pregunta 5

## ¿Qué relación existe entre las variables numéricas?

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes("number").corr(),
    annot=True,
    cmap="coolwarm"
)

plt.show()
```

## Conclusion

El análisis exploratorio permitió identificar patrones relevantes en el comportamiento de los usuarios. Las visualizaciones mostraron diferencias entre planes de suscripción, variaciones según el país y relaciones entre las variables numéricas.
